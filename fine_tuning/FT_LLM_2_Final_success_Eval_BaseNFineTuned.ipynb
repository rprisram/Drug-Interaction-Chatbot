```python
def conv_to_chatml(ex):
  #print(ex)
  msg = ex['messages']
  #print(msg)
  dc = {}
  for indv in msg:
    if indv['role'] == 'user':
      dc['user'] = indv['content']
    elif indv['role'] == 'assistant':
      dc['assistant'] = indv['content']
  return [
          {"role": "user", "content": dc['user']},
          {"role": "assistant", "content": dc['assistant']},
      ]
```

```python
import random
import pandas as pd
import itertools

def augment_and_expand_dataframe(df, num_augmentations=3):
    """
    Expands a DataFrame by creating a fixed number of *unique* augmented examples
    for each row using random sampling of synonym combinations.

    Args:
        df (pd.DataFrame): The original dataframe with pipe-separated synonyms.
        num_augmentations (int): The desired number of new, augmented rows to create
                                 for each original row.

    Returns:
        pd.DataFrame: A new, augmented DataFrame with no duplicate augmentations per original row.
    """
    new_records = []

    for _, row in df.iterrows():
        # Split the pipe-delimited strings into lists of synonyms
        drug1_synonyms = [s.strip() for s in row['Drug1_Name'].split('|')]
        drug2_synonyms = [s.strip() for s in row['Drug2_Name'].split('|')]

        # 1. Generate all possible unique pairs of synonyms (Cartesian product)
        all_possible_pairs = list(itertools.product(drug1_synonyms, drug2_synonyms))

        # 2. Shuffle this list of unique pairs randomly
        random.shuffle(all_possible_pairs)

        # 3. Determine how many unique examples to take
        # This prevents errors if you ask for more augmentations than there are unique pairs
        num_to_generate = min(num_augmentations, len(all_possible_pairs))

        # 4. Take the first 'num_to_generate' from the shuffled list
        for i in range(num_to_generate):
            # Get the unique pair
            selected_drug1, selected_drug2 = all_possible_pairs[i]

            # Create the clean prompt and response
            prompt = f"What is the interaction between {selected_drug1} and {selected_drug2}"
            response = row['Value_Y'].replace("#Drug1", selected_drug1).replace("#Drug2", selected_drug2)

            new_records.append({'prompt': prompt, 'response': response})

    return pd.DataFrame(new_records)
```

```python
def formatting_func_for_map(batch):
    prompts = batch["prompt"]
    responses = batch["response"]
    texts = []

    for prompt, response in zip(prompts, responses):
        conversation = [
            {"role": "user", "content": prompt},
            {"role": "assistant", "content": response},
        ]
        texts.append(tokenizer.apply_chat_template(conversation, tokenize=False, add_generation_prompt=False))

    # The .map() function expects a dictionary where the key is the new column name
    return {"text": texts}
```

```python
def generate_answer(question, max_new_tokens=MAX_NEW_TOKENS_MODEL) -> str:
    """
    Generates an answer using the pre-loaded Unsloth (Hugging Face) model.
    """
    # 1. Format the input using the chat template
    messages = [
        {"role": "user", "content": question},
    ]
    input_ids = base_tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to("cuda")

    # 2. Generate a response using model.generate
    # Note the parameter name changes (e.g., repetition_penalty)
    outputs = base_model.generate(
        input_ids,
        max_new_tokens=max_new_tokens,
        repetition_penalty=1.1,
        temperature=0.1,
        do_sample=True, # Required for temperature to have an effect
        pad_token_id=base_tokenizer.eos_token_id,
    )

    # 3. Decode the output and isolate the answer
    # We decode the full output and then remove the prompt part.
    decoded_output = base_tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    prompt_length = len(base_tokenizer.decode(input_ids[0], skip_special_tokens=True))
    answer = decoded_output[prompt_length:]

    # 4. Manually handle the stop sequence
    if "\n\n" in answer:
        answer = answer.split("\n\n")[0]

    return answer.strip()

# --- Example Usage (remains the same) ---
my_question = "what is the interaction between Prasugrel and Oxpentifylline"
response = generate_answer(my_question)
print(response)
```

```python
def judge_answer(question: str, model_answer: str, ground_truth_answer: str, max_new_tokens=MAX_NEW_TOKENS_JUDGE) -> dict:
    prompt = make_judge_prompt(question, model_answer, ground_truth_answer)
    if judge_tokenizer.pad_token is None:
        judge_tokenizer.pad_token = judge_tokenizer.eos_token
    inputs = judge_tokenizer(
                        prompt,
                        return_tensors="pt",
                        padding=True,
                        truncation=True,
                        max_length=2048,
                        add_special_tokens=True,
                    )
    input_ids = inputs.input_ids.to(device)
    attention_mask = inputs.attention_mask.to(device)
    outputs = judge_model.generate(
        input_ids=input_ids,
        attention_mask = attention_mask,
        max_new_tokens=max_new_tokens,
        temperature=0.5, # Use a low temperature for deterministic evaluation
        top_p=0.95,
        do_sample=False,
        repetition_penalty=1.15,
    )

    # 1. Decode ONLY the newly generated part of the output
    generated_tokens = outputs[0][len(input_ids[0]):]
    decoded = judge_tokenizer.decode(generated_tokens, skip_special_tokens=True)

    # 2. Use a more flexible regex that handles different formats
    #score_match = re.search(r'Score.*?:?\s*(\d+)', decoded, re.IGNORECASE | re.DOTALL) Till 52-101 Eval
    score_match = re.search(r'Final Score:\s*(\d{1,2})', decoded, re.IGNORECASE)

    # Explanation can be the whole decoded text if a specific pattern fails
    #explanation_match = re.search(r"Comparison\\s*:\\s*(.*)", decoded, re.IGNORECASE | re.DOTALL)

    score = int(score_match.group(1)) if score_match else None
    explanation = decoded.strip()

    return {
        "score": score,
        "explanation": explanation,
        "raw_output": decoded,
    }
```

```python
def evaluate_and_save(dataset, question_key, answer_key, output_file=OUTPUT_BASEMODEL_FILE):
    results = []
    # If output file exists, load previous results to resume
    if os.path.exists(output_file):
        print(f"Loading existing results from {output_file} to resume...")
        with open(output_file, "r") as f:
            results = json.load(f)
        processed_questions = {r['question'] for r in results}
    else:
        processed_questions = set()

    for i, example in enumerate(dataset):
        question = example[question_key]
        ground_truth_answer = example[answer_key]

        # Skip if already processed (supports resume)
        if question in processed_questions:
            print(f"Skipping example {i + 1} (already processed).")
            continue

        print(f"\nEvaluating example {i + 1}/{len(dataset)}")
        print(f"Question: {question}")
        print(f"Ground Truth Answer: {ground_truth_answer}")

        # Generate model answer
        try:
            model_answer = generate_answer(question)
        except Exception as e:
            print(f"Error generating answer for question {question}: {e}")
            model_answer = ""

        print(f"Model-generated Answer: {model_answer}")

        # Judge answer
        try:
            judge_result = judge_answer(question, model_answer, ground_truth_answer)
        except Exception as e:
            print(f"Error judging answer for question {question}: {e}")
            judge_result = {"score": None, "explanation": "", "raw_output": ""}

        print(f"Judge Score: {judge_result['score']}")
        print(f"Judge Explanation: {judge_result['explanation']}\n")

        record = {
            "question": question,
            "ground_truth_answer": ground_truth_answer,
            "model_generated_answer": model_answer,
            "judge_score": judge_result["score"],
            "judge_explanation": judge_result["explanation"],
            "judge_raw_output": judge_result["raw_output"],
        }
        results.append(record)

        # Save results incrementally after each example
        with open(output_file, "w") as f:
            json.dump(results, f, indent=2)

    return results
```
